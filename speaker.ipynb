```python
from pyannote.audio import Pipeline

pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization-3.1",
                                    use_auth_token="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")  # Reemplazar con tu token
diarization = pipeline("audio_video.wav")
```

```python
from pyannote.audio import Audio
from pyannote.core import Segment
import torch


#convertimos el audio y la transcriccion  añadiendo los speakers
audio = Audio()
transcript = []

for segment, _, speaker in diarization.itertracks(yield_label=True):
    start = segment.start
    end = segment.end
    segment_text = ""

    for s in segments:
        if s["start"] >= start and s["end"] <= end:
            segment_text += s["text"] + " "

    transcript.append(f"{speaker}: {segment_text.strip()}")

for line in transcript:
    print(line)

```

```python
#se realiza la exportacion del archivo
with open("transcripcion_por_hablante.txt", "w") as f:
    for line in transcript:
        f.write(line + '\n')

files.download("transcripcion_por_hablante.txt")
```
